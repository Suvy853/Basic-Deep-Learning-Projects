```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh activation function
def tanh(x):
    return np.tanh(x)

# ReLU activation function
def relu(x):
    return np.maximum(0, x)
```

```python
# Sample data
x1 = 3
x2 = 2
y_actual = 17

# Weights and biases initialization (random for the start)
np.random.seed(42)  # For reproducibility
weights = {
    'w1': np.random.randn(),  # Weight for x1
    'w2': np.random.randn()   # Weight for x2
}
bias = np.random.randn()  # Bias term

# Define the feed-forward function
def feed_forward(x1, x2, weights, bias):
    # Linear combination
    linear_output = weights['w1'] * x1 + weights['w2'] * x2 + bias
    # Activation function (ReLU for this example)
    activated_output = relu(linear_output)
    return activated_output

# Predict the output
y_pred = feed_forward(x1, x2, weights, bias)

# Display the result
print(f"Predicted Output: {y_pred}")
print(f"Actual Output: {y_actual}")
```

```python
import numpy as np

# Sample data
x1 = 3
x2 = 2
y_actual = 17

# Initialize random weights w1 and w2 between 1 and 10
np.random.seed(42)  # For reproducibility
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

# Predict the output ypred using the equation ypred = x1 * w1 + x2 * w2
y_pred = x1 * w1 + x2 * w2

# Display the initialized weights and predicted output
print(f"Initialized Weight w1: {w1}")
print(f"Initialized Weight w2: {w2}")
print(f"Predicted Output y_pred: {y_pred}")
print(f"Actual Output y_actual: {y_actual}")
```

```python
# Compute the squared error
error = (y_actual - y_pred) ** 2

# Display the squared error
print(f"Squared Error: {error}")
```

```python
# Define the learning rate
learning_rate = 0.01

# Compute the gradients of the error with respect to weights
# Partial derivative of error with respect to w1 and w2
grad_w1 = -2 * (y_actual - y_pred) * x1
grad_w2 = -2 * (y_actual - y_pred) * x2

# Update weights using the gradients and learning rate
w1 = w1 - learning_rate * grad_w1
w2 = w2 - learning_rate * grad_w2

# Display updated weights
print(f"Updated Weight w1: {w1}")
print(f"Updated Weight w2: {w2}")
```

```python
# Initialize random weights
np.random.seed(42)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

# Define the learning rate
learning_rate = 0.01

# Number of iterations
epochs = 20

# Training loop
for epoch in range(epochs):
    # Feed Forward: Predict output
    y_pred = x1 * w1 + x2 * w2

    # Compute squared error
    error = (y_actual - y_pred) ** 2

    # Compute gradients
    grad_w1 = -2 * (y_actual - y_pred) * x1
    grad_w2 = -2 * (y_actual - y_pred) * x2

    # Update weights
    w1 = w1 - learning_rate * grad_w1
    w2 = w2 - learning_rate * grad_w2

    # Print error and weights at each epoch
    print(f"Epoch {epoch + 1}: Error = {error}, w1 = {w1}, w2 = {w2}")

# Final weights and error
print("\nTraining complete.")
print(f"Final Error: {error}")
print(f"Final Weights: w1 = {w1}, w2 = {w2}")
```

```python
import matplotlib.pyplot as plt

# Initialize random weights
np.random.seed(42)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

# Define the learning rate
learning_rate = 0.01

# Number of iterations
epochs = 20

# Lists to store error and predicted output values
error_progression = []
ypred_progression = []

# Training loop
for epoch in range(epochs):
    # Feed Forward: Predict output
    y_pred = x1 * w1 + x2 * w2

    # Compute squared error
    error = (y_actual - y_pred) ** 2

    # Store error and predicted output for visualization
    error_progression.append(error)
    ypred_progression.append(y_pred)

    # Compute gradients
    grad_w1 = -2 * (y_actual - y_pred) * x1
    grad_w2 = -2 * (y_actual - y_pred) * x2

    # Update weights
    w1 = w1 - learning_rate * grad_w1
    w2 = w2 - learning_rate * grad_w2

# Plot the progression of the error across epochs
plt.figure(figsize=(10, 5))
plt.plot(range(1, epochs + 1), error_progression, marker='o', label="Error")
plt.title("Error Progression Across Epochs")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.grid(True)
plt.legend()
plt.show()

# Plot the progression of ypred across epochs
plt.figure(figsize=(10, 5))
plt.plot(range(1, epochs + 1), ypred_progression, marker='o', label="Predicted y")
plt.axhline(y=y_actual, color='r', linestyle='--', label="Actual y")
plt.title("Predicted y Progression Across Epochs")
plt.xlabel("Epochs")
plt.ylabel("Predicted y")
plt.grid(True)
plt.legend()
plt.show()
```
